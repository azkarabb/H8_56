```python
from scipy.stats import uniform
```

```python
n = 10000
start = 10
width = 20
data_uniform = uniform.rvs(size = n, loc = start, scale = width)
```

```python
from scipy.stats import norm

data_normal = norm.rvs(size = 10000, loc = 0, scale = 1)
```

```python
ax = sns.distplot(data_uniform, bins = 100, kde = True, color = 'skyblue', hist_kws = {'linewidth' : 15, 'alpha' : 1})
ax.set(xlabel = 'Uniform Distribution', ylabel = 'Frequency')
```

```python
ax = sns.distplot(data_normal, bins = 100, kde = True, color = 'skyblue', hist_kws = {'linewidth' : 15, 'alpha' : 1})
ax.set(xlabel = 'Uniform Distribution', ylabel = 'Frequency')
```

```python
from scipy.stats import gamma

data_gamma = gamma.rvs(a=5, size = 10000)

ax = sns.distplot(data_gamma,
                 kde = True,
                 bins = 100,
                 color = 'skyblue',
                 hist_kws = {"linewidth" : 15, 'alpha': 1})
ax.set(xlabel = 'Gamma Distribution', ylabel = 'Frequency')
```

```python
from scipy.stats import expon
data_expon = expon.rvs(scale =1, loc = 0, size = 1000)

ax = sns.distplot(data_expon,
                 kde = True,
                 bins = 100,
                 color = 'crimson',
                 hist_kws = {"linewidth" : 15, 'alpha': 1})
ax.set(xlabel = 'Exponetial Distribution', ylabel = 'Frequency')
```

```python
from scipy.stats import poisson

data_poisson = poisson.rvs(mu = 3, size = 10000)

ax = sns.distplot(data_poisson,
                 bins = 30,
                 kde = False,
                 color= 'magenta',
                 hist_kws = {"linewidth" : 15, 'alpha': 1})
ax.set(xlabel = "Poisson Distribution", ylabel = 'Frequency')
```

```python
from scipy.stats import binom
data_binom = binom.rvs(n =10, p = 0.8, size = 10000)

ax = sns.distplot(data_binom,
                 kde = False,
                 color = 'purple',
                 hist_kws = {'linewidth':15, 'alpha' : 1})

ax.set(xlabel = 'Binomial Distribution', ylabel = 'Frequency')
```

```python
import pandas as pd
import numpy as np

df = pd.read_csv('heart.csv')
df
```

```python
df['Sex1'] = df.sex.replace({1: 'Male', 0: 'Female'})
dx = df[['target', 'Sex1']].dropna()
pd.crosstab(dx.target, dx.Sex1)
p_fm = 226/(86+226)
p_fm
```

```python
n = 86 + 226 #226 adalah wanita penderita penyakit jantung, 86 adalah laki-laki.
n #adalah jumlah totalnya
```

```python
se_female = np.sqrt(p_fm * (1 - p_fm) / n)
se_female  
```

```python
#z-score adalah standar statistik dari tabel z, untuk menghitung confidence interval sebanyak 95%.

z_score = 1.96 
lcb = p_fm - z_score * se_female #lower limit of the CI
ucb = p_fm + z_score * se_female #upper limit of the CI

lcb, ucb 
#lcb & ucb dibaca dengan dikalikan dengan 100%
```

```python
import statsmodels.api as sm
sm.stats.proportion_confint(n* p_fm, n)
```
